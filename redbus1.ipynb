```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2

# Database connection parameters
host = "localhost"
port = "5432"
database = "redbus"
username = "postgres"
password = "deva0610"

# Create the connection string
engine_string = f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
engine = create_engine(engine_string)

# Check the connection
try:
    with engine.connect() as connection:
        print("Connection successful!")
except Exception as e:
    print(f"Connection error: {e}")

# Create a sample DataFrame (replace this with your actual DataFrame)
data = {
    'column1': [1, 2, 3],
    'column2': ['A', 'B', 'C']
}
tenstate = pd.DataFrame(data)

# Push data to PostgreSQL
try:
    tenstate.to_sql("redbus", engine, if_exists='append', index=True)
    print("Data successfully pushed to PostgreSQL table!")
except Exception as e:
    print(f"An error occurred: {e}")
```
